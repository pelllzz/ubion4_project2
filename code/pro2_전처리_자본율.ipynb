```python
# TS2000 만능 반복문
# 연결재무제표와 개별재무제표를 비교해서
# 연결재무제표의 값을 우선 채워넣고 개별재무제표의 값을 채워넣습니다
# 둘 다 없다면 결측치 표시를 위한 특정 값을 삽입합니다.
# 다만 전제조건 : df_a, df_b 컬럼이 유사한 데이터 컬럼으로 구성되어야 합니다
# df_a 에 abc 가 있고 df_b에 a`b`c`가 있어야 계산이 제대로 되지
# df_a 에 컬럼이 뭐가 없다던가 df_b 를 살펴보니 컬럼 순서가 뭔가 뒤바뀌어있다
# 이러면 결과를 신뢰할 수 없습니다. 주의하세요.

# csv 한바퀴 돌립니다
for i in range(0,len(df_c)):
    # 컬럼들 순회합니다
    for j in range(0, len(df_c.columns)-3):
        # df_a 에 결측치가 없다면
        if not pd.isna(df_a.iloc[i][j]):
            # df_b 에도 결측치가 없다면
            if not pd.isna(df_b.iloc[i][j]):
                # df_b 삽입
                df_c.iloc[i, j+3] = df_b.iloc[i][j]

            # df_b 가 결측치라면
            else:
                #df_a 값 삽입
                df_c.iloc[i, j+3] = df_a.iloc[i][j]

        # df_a 에 결측치가 있다면
        else:
            # df_b 에 결측치가 없다면
            if not pd.isna(df_b.iloc[i][j]):
                # df_b 삽입
                df_c.iloc[i, j+3] = df_b.iloc[i][j]

            # 둘 다 없으면 결측치로 넣어벼려요
            # 근데 아마 이거 동작 안 할 겁니다. 이미 채워넣었기 때문에...
            else:
                df_c.iloc[i, j+3] = 11111111.11
```
