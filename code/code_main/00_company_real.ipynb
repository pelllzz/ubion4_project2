```python
# 다음 연도의 이자보상배율을 저장
df2['이보배2'] = df2['이자보상배율'].shift(-1).where(df2['회사명']==(df2['회사명']).shift(-1))
df2
```

```python
# 부실징후기업의 경우 1, 아닌 경우 0을 기록

for i in range(0, len(df)):
    if df.iloc[i]['이자보상배율'] < 1:
        df2.loc[i, '부실기업'] = 1
    else:
        df2.loc[i, '부실기업'] = 0
```

```python
df2['부실기업'] = df2['부실기업'].fillna(0)
```
