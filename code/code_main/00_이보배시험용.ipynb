```python
df_b = pd.read_csv("../../raw_data/00_영업손익.csv", thousands=",", encoding="CP949")
df_b
#df_b.drop(columns=["Unnamed: 0"])
```

```python
# 개별재무제표 저장
df_ind = df_b["[제조]* (정상)영업손익(보고서기재)(IFRS)"]

# 연결재무제표 저장
df_oth = df_b["[제조]* (정상)영업손익(보고서기재)(IFRS연결)"]
```

```python
for i in range(0, len(df)):
    # 컬럼들 순회합니다
    if not pd.isna(df_ind.iloc[i]):
        # df_oth 에도 결측치가 없다면
        if not pd.isna(df_oth.iloc[i]):
            # df_oth 삽입
            df.loc[i, "영업손익"] = df_oth.iloc[i]

        # df_oth 가 결측치라면
        else:
            #df_ind 값 삽입
            df.loc[i, "영업손익"] = df_ind.iloc[i]

    # df_ind 에 결측치가 있다면
    else:
        # df_oth 에 결측치가 없다면
        if not pd.isna(df_oth.iloc[i]):
            # df_oth 삽입
            df.loc[i, "영업손익"] = df_oth.iloc[i]

        # 둘 다 결측치면 일단 패스
        else:
            pass
```
