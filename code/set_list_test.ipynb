```python
# 자기자본순이익률 = 당기/자기자본 * 100

df_rate_a = pd.read_csv("../raw_data/자기자본순이익률_raw.csv", encoding="CP949")
```

```python
df_rate_a
```

```python
#AC_con['prev_자산'] = AC_con['자산'].shift().where(AC_con.회사명.eq(AC_con.회사명.shift()))

df_rate_a["[제조]자본(*)(IFRS연결2)"] = df_rate_a["[제조]자본(*)(IFRS연결)"].shift().where(df_rate_a.회사명.eq(df_rate_a.회사명.shift()))
df_rate_a["[제조]자본(*)(IFRS2)"] = df_rate_a["[제조]자본(*)(IFRS)"].shift().where(df_rate_a.회사명.eq(df_rate_a.회사명.shift()))
df_rate_a.head(15)
```

```python
for i in range(0, len(df_rate_a)):
    if df_rate_a.iloc[i]["[제조]자본(*)(IFRS)"] < 0 and df_rate_a.iloc[i]["[제조]당기순이익(손실)(IFRS)"] < 0 :
        df_rate_a.loc[i, "자기자본순이익률(개별)"] = 11111111.11
    else:
        df_rate_a.loc[i, "자기자본순이익률(개별)"] = df_rate_a.iloc[i]["[제조]당기순이익(손실)(IFRS)"]/((df_rate_a.iloc[i]["[제조]자본(*)(IFRS)"] + df_rate_a.iloc[i]["[제조]자본(*)(IFRS2)"])/2)
    
    if df_rate_a.iloc[i]["[제조]자본(*)(IFRS연결)"] < 0 and df_rate_a.iloc[i]["[제조]당기순이익(손실)(IFRS연결)"] < 0 :
        df_rate_a.loc[i, "자기자본순이익률(연결)"] = 11111111.11
    else:
        df_rate_a.loc[i, "자기자본순이익률(연결)"] = df_rate_a.iloc[i]["[제조]당기순이익(손실)(IFRS연결)"]/((df_rate_a.iloc[i]["[제조]자본(*)(IFRS연결)"] + df_rate_a.iloc[i]["[제조]자본(*)(IFRS연결2)"])/2)
```

```python
#df_rate_a.to_csv("../csv_data/자기자본테스트2.csv", encoding="CP949")

kill_2010 = df_rate_a[df_rate_a['회계년도'].str.contains("2010")].index
kill_2010
```

```python
df_rate_a_complete = df_rate_a.drop(kill_2010)
df_rate_a_complete
```
