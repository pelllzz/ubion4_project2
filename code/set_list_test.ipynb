```python
# 위에 있는 columns로 필요한 컬럼 확인하고 복붙하세요
df_test = df_test[['매출액(수익)감소', '매출원가감소', '매출액(수익)감소(연결)','매출원가감소(연결)']]
df_test
```

```python
for i in range(0, len(df_test)):
    if df_test.iloc[i]['매출액(수익)감소'] < 0 and df_test.iloc[i]['매출원가감소'] < 0 :
        df_test.loc[i, '예시1'] = df_test.iloc[i]['매출액(수익)감소'] / df_test.iloc[i]['매출원가감소'] * -1
    # elif df_test.iloc[i]['매출원가감소'] == 0:
    #     df_test.loc[i, '예시1'] = 0
    else:
        df_test.loc[i, '예시1'] = df_test.iloc[i]['매출액(수익)감소'] / df_test.iloc[i]['매출원가감소']
```

```python
# 자기자본순이익률 = 당기/자기자본 * 100

df_rate_a = pd.read_csv("../raw_data/자기자본순이익률_IFRS개별Raw.csv", encoding="CP949")
df_rate_b = pd.read_csv("../raw_data/자기자본순이익률_IFRS연결Raw.csv", encoding="CP949")
```

```python
for i in range(0, len(df_rate_a)):
    if df_rate_a.iloc[i]["[제조]당기순이익(손실)(IFRS)"] < 0 and df_rate_a.iloc[i]["[제조]자본(*)(IFRS)"] < 0 :
        df_rate_a.loc[i, "자기자본순이익률(개별)"] = df_rate_a.iloc[i]["[제조]당기순이익(손실)(IFRS)"] / df_rate_a.iloc[i]["[제조]자본(*)(IFRS)"] * -100
    # elif df_rate_a.iloc[i]['[제조]자본(*)(IFRS)'] == 0:
    #     df_rate_a.loc[i, '예시1'] = 0
    else:
        df_rate_a.loc[i, "자기자본순이익률(개별)"] = df_rate_a.iloc[i]["[제조]당기순이익(손실)(IFRS)"] / df_rate_a.iloc[i]["[제조]자본(*)(IFRS)"] * 100
```
