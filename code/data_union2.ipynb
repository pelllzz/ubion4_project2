```python
df_c = pd.read_csv("../csv_data/hee/직원감소율여부.csv", thousands=",", encoding="CP949")
df_c["직원감소율여부"] = 0
```

```python
df_c.head(1)
```

```python
count = 0

for i in range(0, len(df_c)):
    for j in range(2, len(df_c.columns)-1):
        #print(df_c.iloc[i][j])
        if df_c.iloc[i][j] > 0:
            count += 1

    if count != 0 :
        df_c.loc[i, "직원감소율여부"] = 1
    else :
        df_c.loc[i, "직원감소율여부"] = 0

    count = 0
```

```python
# 총자산대비자본율 여부

df_e = pd.read_csv("../csv_data/hee/총자산대비자본율여부.csv", thousands=",", encoding="CP949" )
df_e
```
