```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./checking2.csv", thousands=",", encoding="CP949")
df.drop(columns=["Unnamed: 0", "count"], inplace=True)
```

```python
df['부실기업'] = df['부실기업'].fillna(0)
```

```python
df
```

```python
df['이보배2'] = df['이자보상배율'].shift(-1).where(df['회사명']==(df['회사명']).shift(-1))
df
```

```python
df.head(30)
```

```python
df2 = df[['회사명', '거래소코드', '회계년도', '이자보상배율', '이보배2', '부실기업']]
df2
```

```python
df2.head(15)
```

```python
for i in range(0, len(df2)):
    if df2.iloc[i]['이자보상배율'] < 1 and df2.iloc[i]['이보배2'] > 1:
        df2.loc[i, '회생여부'] = 1
```

```python
df2.to_csv("checking4.csv", encoding="CP949")
```
