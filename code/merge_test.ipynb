```python
#AC_con['prev_자산'] = AC_con['자산'].shift().where(AC_con.회사명.eq(AC_con.회사명.shift()))

df_c['매출액(수익)감소'] = df_c['매출액(수익)(*)(IFRS)(천원)'].shift().where(df_c.회사명.eq(df_c.회사명.shift()))
df_c['매출원가감소'] = df_c['매출원가(*)(IFRS)(천원)'].shift().where(df_c.회사명.eq(df_c.회사명.shift()))
df_c['매출액(수익)감소(연결)'] = df_c['매출액(수익)(*)(IFRS연결)(천원)'].shift().where(df_c.회사명.eq(df_c.회사명.shift()))
df_c['매출원가감소(연결)'] = df_c['매출원가(*)(IFRS연결)(천원)'].shift().where(df_c.회사명.eq(df_c.회사명.shift()))
```

```python
df_c['매출액(수익)감소'] = (df_c['매출액(수익)감소'] - df_c['매출액(수익)(*)(IFRS)(천원)'])/df_c['매출액(수익)(*)(IFRS)(천원)']
df_c['매출원가감소'] = (df_c['매출원가감소'] - df_c['매출원가(*)(IFRS)(천원)'])/df_c['매출원가(*)(IFRS)(천원)']
df_c['매출액(수익)감소(연결)'] = (df_c['매출액(수익)감소(연결)'] - df_c['매출액(수익)(*)(IFRS연결)(천원)'])/df_c['매출액(수익)(*)(IFRS연결)(천원)']
df_c['매출원가감소(연결)'] = (df_c['매출원가감소(연결)'] - df_c['매출원가(*)(IFRS연결)(천원)'])/df_c['매출원가(*)(IFRS연결)(천원)']
```
